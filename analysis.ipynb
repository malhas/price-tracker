## Import Modules

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

## Load CSVs

```python
cybex_data = pd.read_csv("items/cadeira-cybex.csv")
cafe_data = pd.read_csv("items/cafe-masaka.csv")
jeans_data = pd.read_csv("items/jeans.csv")
lentes180_data = pd.read_csv("items/lentes30.csv")
lentes180_data = pd.read_csv("items/lentes90.csv")
lentes180_data = pd.read_csv("items/lentes180.csv")
```

## Analyze Data

### Cadeira Cybex

```python
cybex_data.tail()
```

```python
pd.options.display.float_format = '{:.2f}'.format
years = mdates.YearLocator()
months = mdates.MonthLocator()
days = mdates.DayLocator()
date_fmt = mdates.DateFormatter("%d-%m-%Y")
cybex_data.Date = pd.to_datetime(cybex_data.Date, format="%d-%m-%Y").dt.strftime("%d-%m-%Y")
cybex_data.Date = pd.to_datetime(cybex_data.Date, format="%d-%m-%Y")

plt.figure(figsize=(6,3), dpi=120)
plt.title('Cadeira Cybex Price', fontsize=12)
# Increase the size and rotate the labels on the x-axis
plt.xticks(fontsize=8, rotation=45)
ax1 = plt.gca()
ax2 = ax1.twinx()
ax1.set_ylabel('Cadeira Price', color='#E6232E', fontsize=12)
ax2.set_ylabel('SMA Trend', color='skyblue', fontsize=12)
# Set the minimum and maximum values on the axes
ax1.set_ylim([0, 300])
ax2.set_ylim([0, 300])
ax1.set_xlim([cybex_data.Date.min(), cybex_data.Date.max()])
ax1.plot(cybex_data.Date, cybex_data.Price, color='#E6232E', linewidth=3)
ax2.plot(cybex_data.Date, cybex_data.Price.rolling(window=6).mean(), color='skyblue', linewidth=3, marker='o')
ax1.xaxis.set_major_formatter(date_fmt)
plt.show()
```
